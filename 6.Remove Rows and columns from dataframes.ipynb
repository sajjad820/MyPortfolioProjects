```python
student = {'first_name': ['alter','shirin','qiqi','ozge','hisam'],
           'last_name':['gudu','mony','tai','bisgin','gudu'],
           'email':['alter_gudu12@gmail.com','shirinmony@gmail.com','qiqi34@yahoo.com','bisginozge@gmail.com','hismgudu@gmail.com'],
           'matri. num':['10243','50678','76545','32145','67856']}
```

```python
import pandas as pd
```

```python
student_ds= pd.DataFrame(student)
student_ds
```

Adding two columns

```python
student_ds['full_name']= student_ds['first_name']+' '+student_ds['last_name']
student_ds
```

Dropping two columns using 'drop' function

```python
student_ds.drop(columns=['first_name','last_name'], inplace= True) #When we use 'inplace=True' the change is applied to the original dataset 
student_ds
```

Spliting out the string values of a column using 'split' funciton

```python
student_ds['full_name'].str.split(' ', expand=True)
```

```python
student_ds[['first_name','last_name']]= student_ds['full_name'].str.split(' ', expand=True) #use of split function using column name 
student_ds
```

Inserting a value to a column using 'append' funtion

```python
student_ds.append({'first_name':'tony'},ignore_index=True)
```

```python
student_2 = {'first_name': ['qiara','Ben'],
           'last_name':['advani','girnus'],
           'email':['advani_qiara42@gmail.com','ben_girnus@gmail.com'],
           'matri. num':['67890','98342']}
```
