```python
student = {'first': ['alter','shirin','qiqi','ozge','hisam'],
           'last':['gudu','mony','tai','bisgin','gudu'],
           'email':['alter_gudu12@gmail.com','shirinmony@gmail.com','qiqi34@yahoo.com','bisginozge@gmail.com','hismgudu@gmail.com'],
           'matri. num':['10243','50678','76545','32145','67856']}
```

```python
import pandas as pd
```

```python
student_ds= pd.DataFrame(student)
student_ds
```

```python
student_ds.columns = ['first name', 'last name', 'email', 'matri.num'] #How to change the column name in dataframe
student_ds
```

```python
student_ds.columns= [x.upper() for x in student_ds.columns] # to change the column name in capital letters 
student_ds.columns= [x.lower() for x in student_ds.columns] # to change the column name in small letters 
student_ds
```

```python
student_ds.columns= student_ds.columns.str.replace(' ','_')
student_ds.columns= student_ds.columns.str.replace('.','_') #changing the column in terms of specific criteria
student_ds
```

```python
student_ds.rename(columns={'first_name':'first','last_name':'last'}) #another way of manipulatin the specific columns names
```

```python
student_ds.loc[4] = ['hisam','guda','hisamguda@gmail.com','67856'] #changing the information of all the columns of a specific row 
student_ds
```

```python
student_ds.loc[4,'matri_num']='67856' #for changing the info of a one column
#OR
student_ds.at[4,'matri_num']='67856' #another option changing the info of a one column
student_ds.loc[4,['matri_num','last_name',]]='67856','guda'#for changing the info for multiple columns 
student_ds
```

```python
student_ds['email']= student_ds['email'].str.upper() #changing the letter format of a specific columns for all rows
#OR
student_ds['email']= student_ds['email'].str.lower()
student_ds
```

```python
student_ds['email'].apply(len) #to apply a function on a specific column for all the rows
```

```python
def update_email(email): 
    return email.upper()
```

```python
student_ds['email'].apply(update_email)
```

```python
student_ds.apply(len,axis='rows') #how many rows in each column 
student_ds.apply(len,axis='columns')#how many columns in each row 
```
