How do filter works in dataset?

Getting the dataframe based on a specific condition 

```python
filt= (student_df['last_name']=='gudu') #this code will show only boolean value 
#filt
student_df.loc[filt] #this code will show all the information in the dataframe in terms of condition 
#OR
student_df.loc[filt,'email'] #it will show information only from 'email' column that meets the given condiion to filter
```

Filtering the dataframe wih multiple conditions 

```python
filt1= (student_df['last_name']=='gudu') |(student_df['first_name']=='qiqi') #it shows only boolean values
#filt1
#OR
student_df.loc[~filt1,'first_name'] #it shows the information from 'first_name' column from dataframe which does not match with 'filt1' condition
```
