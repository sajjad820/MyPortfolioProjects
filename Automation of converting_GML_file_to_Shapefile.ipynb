# **Steps for Automation using 'For' loop function (for many files)**

```python
file_folder= '/content/drive/MyDrive/Green Adapt/3d-gm_lod1_kacheln' 

for filenames in os.walk(file_folder):
  all_files=glob.glob(os.path.join(file_folder,'*.gml')) # making a list of all the paths of gml files
```

```python
x= np.array(all_files) #creating an array to get index numbers for every file to run 'for' loop' 
```

```python
for i in x:
  ds_conversion=gdal.VectorTranslate(i+'.shp',i,
                       options='-f "ESRI Shapefile"') #gdal.Vectortranslate function
  ds= gpd.read_file(i+'.shp', driver= 'ESRI Shapefile')
  ds= ds.set_crs('EPSG:25832')
  ds= ds.to_file(i+'.shp')

```
