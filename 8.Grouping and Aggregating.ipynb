```python
python_user_country= country_grp['LanguageHaveWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
python_user_country
```

```python
respondent_per_country= ds['Country'].value_counts()
respondent_per_country
```

```python
python_df= pd.concat([respondent_per_country,python_user_country],axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'LanguageHaveWorkedWith':'Python_user','Country':'NumOfRespondents'}, inplace= True)
python_df
```

```python
python_df['percentageOfUser']=(python_df['Python_user']/python_df['NumOfRespondents'])*100 
python_df
```

```python
python_df.sort_values(by='percentageOfUser',ascending=False, inplace=True)
python_df.head(50)
```
